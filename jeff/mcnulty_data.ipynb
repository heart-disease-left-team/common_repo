```python
# set columns of interest
fields = ['num','age','sex','chol','smoke','cigs','years','dm','famhist']

# replace nans with zero (categorical) or avg value (numerical)
subdf = hd[fields]
subdf.chol = subdf.chol.fillna(subdf.chol.mean())
subdf = subdf.fillna(0)
subdf.num[subdf.num > 0] = 1

subdf['cholbin'] = subdf.chol
subdf.cholbin[subdf.cholbin<200] = 0
subdf.cholbin[subdf.cholbin>=200] = 1
subdf = subdf.drop('chol',1)

subdf['cigsbin'] = subdf.cigs
subdf.cigsbin[subdf.cigsbin>0] = 1
subdf = subdf.drop('cigs',1)
```

```python
smalldf = subdf.drop(['years','famhist'],1)
smalldf['age25_34'] = (smalldf.age <= 34)*1
smalldf['age35_44'] = ((smalldf.age >= 35) & (smalldf.age <= 44))*1
smalldf['age45_54'] = ((smalldf.age >= 45) & (smalldf.age <= 54))*1
smalldf['age55_64'] = ((smalldf.age >= 55) & (smalldf.age <= 64))*1
smalldf['age65'] = (smalldf.age >= 65)*1


smalldf = smalldf.drop('age',1)
smalldf.head()
```

```python
scores = []
precision = []
recall = []
Crange = np.linspace(0.1,2,50)
for val in Crange:
    #model = SVC(C=val)
    model = LogisticRegression(penalty='l1',C=val)
    model.fit(Xtrain,Ytrain)
    Ypred = model.predict(Xtest)
    scores.append(accuracy_score(Ytest,Ypred))
    precision.append(precision_score(Ytest,Ypred))
    recall.append(recall_score(Ytest,Ypred,average=None,pos_label=1)[0])

ppl.plot(Crange,scores,label='accuracy')
ppl.plot(Crange,precision,label='precision')
ppl.plot(Crange,recall,label='recall')
plt.xlabel('C value')
plt.title('model performance')
plt.legend(loc='best')
```
